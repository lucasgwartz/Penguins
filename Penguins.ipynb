```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plp
%matplotlib inline
```

```python
print(sns.get_dataset_names())
```

The Penguins Dataset came from Seaborn.

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins
```

## Analyzing Dataset

```python
sns.set_theme(style="whitegrid", palette="Set3")
sns.countplot(x='sex',data=penguins)
```

```python
sns.set_theme(style="whitegrid",palette="Paired")
sns.countplot(x='island',data=penguins)
```

```python
sns.set_theme(style="whitegrid", palette="vlag")
sns.pairplot(penguins)
```

```python
sns.set_theme(style="whitegrid", palette="colorblind")
sns.swarmplot(data=penguins,x='body_mass_g',y='sex',hue='species')
```

```python
sns.set_theme(style="whitegrid", palette="deep")
sns.violinplot(y=penguins['bill_length_mm'],x=penguins['species'])
```

```python
sns.set_theme(style="whitegrid", palette="bright")
sns.stripplot(y = penguins['bill_depth_mm'],x=penguins['species'])
```

```python
sns.set_theme(style="whitegrid", palette="pastel")
sns.barplot(y=penguins['flipper_length_mm'],x=penguins['species'])
```
